### A reminder: the math behind the model

To refresh your memory, the equation for the height of the ball is given by:

$$h(t) = h_0 + v_0t - \frac{1}{2}gt^2$$

where $h$ is the height of the ball as a function of time, $h_0$ is the initial height that the ball is launched from, $v_0$ is the initial speed the ball is launched with, $g$ is the acceleration due to gravity, and $t$ is the time in seconds. 

As before, you know that the initial height that the ball is launched from, $h_0$, is roughly your height since you'll be throwing it, which we'll approximate as **1.75 meters**.

You also recall that at some point you learned that the **acceleration due to gravity is 9.8 m/s$^2$**

### So, how high will the ball go and when will it get there?

1. Starting at $t = 0$, use a loop to compute the height of the ball as a function of time until $t = 10$, in increments of 0.5 seconds.
2. Using an `if` statement, determine the maximum height of the ball and the time at which the ball reaches that height. At the end of the loop, print out the maximum height and the corresponding time.
3. Using an `if` or `elif` statement, when the height of the ball starts to decrease, print "The ball is falling!".

**Work out a plan on your whiteboard before you start coding!**

```python
# Put your code here

t = []
for i in range(20+1):
    t.append(i/2)

h=1.75
g=9.8
speed=47
    
height=[]
for i in range(len(t)):
    #for j in range(len(speeds)):
    height.append(h+speed*t[i] -0.5*g*t[i]**2)
height_max=0;    
for i in range(len(height)):
    
    if height_max<height[i]:
        height_max=height[i]
        time_max=t[i]
    elif height_max>height[i]:
        print('the ball is falling','max height',height_max,'time',time_max)
        
```
